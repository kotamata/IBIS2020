```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline

pd.set_option('display.max_columns', None)
```

```python
data = pd.read_table('german.data', header=None, sep=' ')
column = ["checking_account", "duration", "credit_history", "purpose", "credit_amount", "saving_account", 
          "employment", "installment_rete", "sex", "debtors", "residence", "property", "age", "installment_plans", 
          "housing", "credit", "job", "people", "telephone", "foreign_work", "label"]
data.columns = column
data.head()
```

orderd categorical : checking account, job, saving account, employment, property

categorical : Sex, housing, purpose, credit_history, debtors, installment plans, housing, telephone, 

numrical : duration, age, Credit amount, installment rate, residence

```python
data = pd.read_table('german.data', header=None, sep=' ')
data.columns = column
#orderd categorical feature
data["checking_account"] = data["checking_account"].map( {'A11': 1 , 'A12':2 , 'A13':3, 'A14':0}).astype(int)
data["credit_history"] = data["credit_history"].map( {'A30': 0 , 'A31':1 , 'A32':2, 'A33':3, 'A34':4}).astype(int)
data["saving_account"] = data["saving_account"].map( {'A61': 1 , 'A62':2, 'A63':3, 'A64':4, 'A65':0}).astype(int)
data["employment"] = data["employment"].map( {'A71': 0 , 'A72':1 , 'A73':2, 'A74':3, 'A75':4}).astype(int)
data["property"] = data["property"].map( {'A121': 0 , 'A122':1 , 'A123':2, 'A124':4}).astype(int)
data["job"] = data["job"].map( {'A171': 0 , 'A172':1 , 'A173':2, 'A174':3}).astype(int)

#categorical feature
#data["purpose"] = data["purpose"].map( {'A40': 0 , 'A41':0 , 'A43':1})
data["purpose"] = data["purpose"].apply( lambda x : "car" if x=='A40' or x=='A41' else "radio/TV" if x=='A43' else "others")
#data["purpose"] = data["purpose"].apply(lambda x: 2 if x == np.nan else None)
data["sex"] = data["sex"].map( {'A91': 0 , 'A92':1 , 'A93':0, 'A94':0, 'A95':1}).astype(int)
data["debtors"] = data["debtors"].map( {'A101': "none", 'A102':"co-applicant" , 'A103':"guarantors"})
data["installment_plans"] = data["installment_plans"].map( {'A141':"bank" , 'A142':"stores", 'A143':"none"})
data["housing"] = data["housing"].map( {'A151': "rent", 'A152':"own" , 'A153':"for free"})
data["telephone"] = data["telephone"].map( {'A191': 0 , 'A192':1}).astype(int)
data["foreign_work"] = data["foreign_work"].map( {'A201': 0 , 'A202':1}).astype(int)

data["label"] -= 1

#data['Sex'] = data['Sex'].apply(lambda x: 1 if x == 'male' else 0)
#df['Embarked'] = df['Embarked'].map( {'S': 0 , 'C':1 , 'Q':2}).astype(int)

print("one hot encoding for purpose, debtors, installment_plans, housing")
```

```python
#one hot encoding for purpose, debtors, installment_plans, housing
categorical = ["purpose", "debtors", "installment_plans", "housing"]
encode_data = pd.get_dummies(data, columns=categorical)
encode_data
encode_data.head()
```

```python
drop_feature = ["saving_account", "employment", "debtors_co-applicant", "debtors_guarantors", 
                "debtors_none", "residence", "credit", "job", "people", "telephone", "foreign_work", "purpose_car", "purpose_radio/TV", 
                "purpose_others", "housing_rent", "housing_for free", "installment_plans_bank", "installment_plans_none", "installment_plans_stores"]
drop_data = encode_data.drop(drop_feature, axis=1)
drop_data.head()
```
