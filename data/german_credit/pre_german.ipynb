```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline

pd.set_option('display.max_columns', None)
```

```python
data = pd.read_table('german.data', header=None, sep=' ')
data = data.rename(columns={20 : 'label'})
data.head()
```

```python
print("shape of data")
print(data.shape)
print()
print("number of unique")
print(data.nunique())
```

```python
print(data.isnull().sum())
```

```python
categorical = [0, 2, 3, 5, 6, 8, 9, 11, 13, 14, 16, 18, 19]
encode_data = pd.get_dummies(data, columns=categorical)
encode_data
encode_data.head()
```

```python
print("shape of data")
print(encode_data.shape)
```

```python
encode_data.to_csv("pre_german.csv")
```
